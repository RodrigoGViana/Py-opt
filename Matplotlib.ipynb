```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xs = np.linspace(1,500,1000)
ys = 10 * xs ** 2 + 5
```

```python
plt.plot(xs, ys)
plt.show()
```

```python
plt.plot(xs, ys)
plt.title('Gráfico de la función $10x^2 + 5$')
plt.xlabel('Variable independiente')
plt.ylabel('Variable dependiente')
plt.show()
```

```python
plt.plot(xs, ys)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.show()
```

```python
a = np.linspace(-10, 10, 500)
b = np.cos(a)
c = np.sin(a)
```

```python
plt.plot(a, b,'-k', a, c, '--k')
plt.legend(('línea 1', 'línea 2'), loc='lower right')
plt.show()
```

```python
x = np.linspace(-50,50,20000)
ecuacion1 = 0.5 - 2.5 * x
ecuacion2 = 5/3 + x

plt.plot(x, ecuacion1, '-r', linewidth=2.0, label='Ecuación 1')
plt.plot(x, ecuacion2, '-b', linewidth=2.0, label='Ecuación 2')
plt.legend()
plt.show()
```
