```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xs = np.linspace(1,500,1000)
ys = 10 * xs ** 2 + 5
```

```python
plt.plot(xs, ys)
plt.show()
```

```python
plt.plot(xs, ys)
plt.title('Gr치fico de la funci칩n $10x^2 + 5$')
plt.xlabel('Variable independiente')
plt.ylabel('Variable dependiente')
plt.show()
```

```python
plt.plot(xs, ys)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.show()
```

```python
a = np.linspace(-10, 10, 500)
b = np.cos(a)
c = np.sin(a)
```

```python
plt.plot(a, b,'-k', a, c, '--k')
plt.legend(('l칤nea 1', 'l칤nea 2'), loc='lower right')
plt.show()
```
